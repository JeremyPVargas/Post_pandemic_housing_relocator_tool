```python
import requests

url = "https://zillow-com1.p.rapidapi.com/propertyExtendedSearch"

querystring = {"location":"atlanta, ga" or "austin, tx", "status_type":"ForRent", "price":""}

headers = {
	"X-RapidAPI-Key": rapid_api_key,
	"X-RapidAPI-Host": "zillow-com1.p.rapidapi.com"
}

z_rent = requests.request("GET", url, headers=headers, params=querystring)
```

```python
z_rent
```

```python
soup= BeautifulSoup(z_rent.content, 'html.parser')
```

```python
z_rent_json = z_rent.json()
```

```python
z_rent_json
```

```python
zillow_df = pd.json_normalize(data=z_rent_json['props'])
```
